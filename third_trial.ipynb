```python
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
import time
```

```python
mentor = pd.read_csv('mentor.csv')
mentee = pd.read_csv('mentee.csv')

# Mentor ve mentee veri setlerini birleştirme
merged_df = pd.merge(mentee, mentor, left_on='Mentor_ID', right_on='ID', suffixes=('_mentee', '_mentor'))

# Gerekli sütunları seçme
merged_df = merged_df[['ID_mentee', 'Category_mentee', 'Interests_mentee', 'Rating', 'ID_mentor', 'Category_mentor', 'Interests_mentor', 'Name_mentor', 'Surname_mentor']]

# Sütunları yeniden adlandırma
merged_df = merged_df.rename(columns={'ID_mentee': 'Mentee_ID', 'Category_mentee': 'Mentee_Category', 'Interests_mentee': 'Mentee_Interests',
                                      'ID_mentor': 'Mentor_ID', 'Category_mentor': 'Mentor_Category', 'Interests_mentor': 'Mentor_Interests',
                                      'Name_mentor': 'Mentor_Name', 'Surname_mentor': 'Mentor_Surname'})

# İlk 5 satırı gösterme
merged_df.head()
```

```python
# Mentor veri setinde eksik değerleri kontrol etme
print(merged_df.isnull().sum())
```

```python
# Ilgi alanlarını ayrı öğeler olarak ele alarak her bir kullanıcı-öğe matrisini oluşturma
interests_mentee = set()
interests_mentor = set()

for interest_list in merged_df['Mentee_Interests']:
    interests_mentee.update(interest_list.split(", "))

for interest_list in merged_df['Mentor_Interests']:
    interests_mentor.update(interest_list.split(", "))

interests_mentee = list(interests_mentee)
interests_mentor = list(interests_mentor)

interests_index_mentee = {interest: idx for idx, interest in enumerate(interests_mentee)}
interests_index_mentor = {interest: idx for idx, interest in enumerate(interests_mentor)}

def encode_interests(interest_list, interests_index):
    encoded = np.zeros(len(interests_index))
    for interest in interest_list.split(", "):
        if interest in interests_index:
            encoded[interests_index[interest]] = 1
    return encoded

merged_df['encoded_mentee_interests'] = merged_df['Mentee_Interests'].apply(lambda x: encode_interests(x, interests_index_mentee))
merged_df['encoded_mentor_interests'] = merged_df['Mentor_Interests'].apply(lambda x: encode_interests(x, interests_index_mentor))
```

```python
# Eğitim ve test setlerine ayırma
train_df, test_df = train_test_split(merged_df, test_size=0.2, random_state=42)

# NMF modelini başlatma ve uygulama
n_components = 5  # Faktör sayısı
model = NMF(n_components=n_components, init='random', random_state=0, max_iter=200)  # Epoch sayısı 200 olarak belirlendi

# Eğitim seti ile modelin eğitilmesi
start_time = time.time()
mentee_factors_train = model.fit_transform(np.array(train_df['encoded_mentee_interests'].tolist()))
mentor_factors_train = model.fit_transform(np.array(train_df['encoded_mentor_interests'].tolist()))
end_time = time.time()
training_time = end_time - start_time

# Yeniden yapılandırma hatalarını hesaplama
train_reconstruction_error = model.reconstruction_err_

# Test seti için faktörlerin çıkarılması
mentee_factors_test = model.transform(np.array(test_df['encoded_mentee_interests'].tolist()))
mentor_factors_test = model.transform(np.array(test_df['encoded_mentor_interests'].tolist()))

# Test seti yeniden yapılandırma hatası
test_reconstruction_error = np.linalg.norm(np.array(test_df['encoded_mentee_interests'].tolist()) - np.dot(mentee_factors_test, model.components_))

# Eğitim süresi ve hataları görüntüleme
print(f"Training time: {training_time} seconds")
print(f"Training reconstruction error: {train_reconstruction_error}")
print(f"Test reconstruction error: {test_reconstruction_error}")

```

```python
def get_recommendations(new_user_interests, model, train_df, interests_index_mentor, N=10):
    # Yeni kullanıcının ilgi alanlarını encode etme
    encoded_interests = encode_interests(new_user_interests, interests_index_mentor)
    new_user_factors = model.transform([encoded_interests])
    
    # Tüm mentorların faktörlerini çıkarma
    mentor_factors = model.transform(np.array(train_df['encoded_mentor_interests'].tolist()))
    
    # Cosine benzerliğini hesaplama
    similarities = cosine_similarity(new_user_factors, mentor_factors)[0]
    
    # Benzerlik puanlarına göre mentorları sıralama
    mentor_indices = np.argsort(similarities)[::-1]
    
    recommended_mentors = []
    
    # Mentorları benzerlik puanına göre sırayla kontrol et
    for mentor_index in mentor_indices[:N]:
        mentor_id = train_df.iloc[mentor_index]['Mentor_ID']
        mentor_name = train_df.iloc[mentor_index]['Mentor_Name']
        mentor_surname = train_df.iloc[mentor_index]['Mentor_Surname']
        mentor_interests = train_df.iloc[mentor_index]['Mentor_Interests']
        recommended_mentors.append({'Mentor_ID': mentor_id, 'Mentor_Name': mentor_name, 'Mentor_Surname': mentor_surname, 'Mentor_Interests': mentor_interests})
    
    # DataFrame'e dönüştürme
    recommendations_df = pd.DataFrame(recommended_mentors)
    return recommendations_df

# Yeni bir mentee için öneri
new_mentee_interests = "Makine Öğrenmesi, Yapay Zeka, Yazılım Geliştirme"
recommendations = get_recommendations(new_mentee_interests, model, train_df, interests_index_mentor)
print(recommendations)
```
