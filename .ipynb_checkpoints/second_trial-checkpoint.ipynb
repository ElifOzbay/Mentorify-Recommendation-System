```python
# Her bir ilgi alanı altında benzersiz subinterestleri bulma
interest_subinterests = {}
for index, row in mentor.iterrows():
    category = row['Category']
    interests = row['Interests'].split(", ")
    if category not in interest_subinterests:
        interest_subinterests[category] = set()
    interest_subinterests[category].update(interests)

# Her bir ilgi alanı altında benzersiz subinterestlerin sayısını hesaplama
interest_counts = {}
for category, subinterest_set in interest_subinterests.items():
    interest_counts[category] = {}
    for subinterest in subinterest_set:
        count = mentor['Interests'].apply(lambda x: subinterest in x).sum()
        interest_counts[category][subinterest] = count

# Grafik için veri hazırlığı ve çubuk grafik oluşturma
num_categories = len(interest_counts)
num_rows = num_categories // 3 if num_categories % 3 == 0 else num_categories // 3 + 1
fig, axs = plt.subplots(num_rows, 3, figsize=(16, 4 * num_rows))  # Numara satırı 3 olduğu sürece burası değişebilir
for i, (category, subinterest_count) in enumerate(interest_counts.items()):
    ax = axs[i // 3, i % 3]  # i değerine göre uygun alt grafiğe erişim sağlıyor
    ax.bar(subinterest_count.keys(), subinterest_count.values(), color='purple')
    ax.set_title(f'{category}', fontsize=12)
    ax.set_ylabel('Mentor Sayısı', fontsize=10)
    ax.tick_params(axis='x', rotation=90)  # x ekseninde etiketleri döndürme
plt.tight_layout()  # Grafikler arasındaki boşluğu otomatik olarak ayarlıyor
plt.show()
```

```python
import numpy as np
import pandas as pd

# Mentor ve mentee veri setlerini yükleme
mentor = pd.read_csv('mentor.csv')
mentee = pd.read_csv('mentee.csv')

# Mentor ve mentee veri setlerini gerekli sütunları seçerek düzenleme
mentor = mentor[['ID', 'Category', 'Interests']]
mentee = mentee[['ID', 'Category', 'Interests', 'Rating', 'Mentor_ID']]

# Mentor ve mentee veri setlerini birleştirme
merged_df = pd.merge(mentor, mentee, on='Category')

# Kullanıcı-öğe matrisini oluşturma
ratings_matrix = merged_df.pivot(index='Mentee_ID', columns='Mentor_ID', values='Rating')

# NaN değerleri 0 ile doldurma
ratings_matrix = ratings_matrix.fillna(0)

# Kullanıcı-öğe matrisini numpy dizisine dönüştürme
R = ratings_matrix.to_numpy()

# Kullanıcı ve öğe sayısını almak
num_users, num_items = R.shape

# Faktör sayısını belirleme
K = 5

# Kullanıcı ve öğe faktör matrislerini rastgele başlatma
P = np.random.normal(scale=1./K, size=(num_users, K))
Q = np.random.normal(scale=1./K, size=(num_items, K))

# Öğrenme oranı ve regülarizasyon terimlerini ayarlama
alpha = 0.01
lambda_ = 0.02

# Eğitim döngüsü
for step in range(1000):
    for i in range(num_users):
        for j in range(num_items):
            if R[i, j] > 0:
                # Hata hesaplama
                eij = R[i, j] - np.dot(P[i, :], Q[j, :].T)
                # Gradient descent ile P ve Q'yu güncelleme
                P[i, :] = P[i, :] + alpha * (2 * eij * Q[j, :] - lambda_ * P[i, :])
                Q[j, :] = Q[j, :] + alpha * (2 * eij * P[i, :] - lambda_ * Q[j, :])

# Tahmin matrisini hesaplama
nR = np.dot(P, Q.T)

# Tahminleri DataFrame'e dönüştürme
predictions_df = pd.DataFrame(nR, index=ratings_matrix.index, columns=ratings_matrix.columns)

# Bir mentee için önerileri almak
def recommend_mentor(mentee_id, predictions_df, num_recommendations=5):
    # Mentee'nin derecelendirmelerini al
    user_row = predictions_df.loc[mentee_id]
    # Zaten derecelendirilmiş mentorları kaldır
    user_row = user_row[user_row == 0]
    # En yüksek tahminli mentorları al
    top_recommendations = user_row.sort_values(ascending=False).head(num_recommendations)
    return top_recommendations.index.tolist()

# Örnek bir mentee için mentor önerileri al
recommendations = recommend_mentor(1, predictions_df)
print(f"Mentee 1 için önerilen mentorlar: {recommendations}")
```

```python
# Mentor ve mentee veri setlerini yükleme
mentor = pd.read_csv('mentor.csv')
mentee = pd.read_csv('mentee.csv')

# Mentor ve mentee veri setlerini birleştirme
merged_df = pd.merge(mentee, mentor, left_on='Mentor_ID', right_on='ID', suffixes=('_mentee', '_mentor'))

# Gerekli sütunları seçme
merged_df = merged_df[['ID_mentee', 'Category_mentee', 'Interests_mentee', 'Rating', 'ID_mentor', 'Category_mentor', 'Interests_mentor']]

# Sütunları yeniden adlandırma
merged_df = merged_df.rename(columns={'ID_mentee': 'Mentee_ID', 'Category_mentee': 'Mentee_Category', 'Interests_mentee': 'Mentee_Interests',
                                      'ID_mentor': 'Mentor_ID', 'Category_mentor': 'Mentor_Category', 'Interests_mentor': 'Mentor_Interests'})

# İlk 5 satırı gösterme
merged_df.head()
```

```python
# Kullanıcı-öğe matrisini oluşturma
ratings_matrix = merged_df.pivot(index='Mentee_ID', columns='Mentor_ID', values='Rating')

# NaN değerleri 0 ile doldurma
ratings_matrix = ratings_matrix.fillna(0)

# Kullanıcı-öğe matrisini numpy dizisine dönüştürme
R = ratings_matrix.to_numpy()

# Kullanıcı ve öğe sayısını almak
num_users, num_items = R.shape

# Faktör sayısını belirleme
K = 5

# Kullanıcı ve öğe faktör matrislerini rastgele başlatma
P = np.random.normal(scale=1./K, size=(num_users, K))
Q = np.random.normal(scale=1./K, size=(num_items, K))

# Öğrenme oranı ve regülarizasyon terimlerini ayarlama
alpha = 0.01
lambda_ = 0.02

# Eğitim döngüsü
for step in range(1000):
    for i in range(num_users):
        for j in range(num_items):
            if R[i, j] > 0:
                # Hata hesaplama
                eij = R[i, j] - np.dot(P[i, :], Q[j, :].T)
                # Gradient descent ile P ve Q'yu güncelleme
                P[i, :] = P[i, :] + alpha * (2 * eij * Q[j, :] - lambda_ * P[i, :])
                Q[j, :] = Q[j, :] + alpha * (2 * eij * P[i, :] - lambda_ * Q[j, :])

# Tahmin matrisini hesaplama
nR = np.dot(P, Q.T)

# Tahminleri DataFrame'e dönüştürme
predictions_df = pd.DataFrame(nR, index=ratings_matrix.index, columns=ratings_matrix.columns)

```

```python
# Bir mentee için önerileri almak
def recommend_mentor(mentee_id, predictions_df, num_recommendations=5):
    # Mentee'nin derecelendirmelerini al
    user_row = predictions_df.loc[mentee_id]
    # Zaten derecelendirilmiş mentorları kaldır
    user_row = user_row[user_row == 0]
    # En yüksek tahminli mentorları al
    top_recommendations = user_row.sort_values(ascending=False).head(num_recommendations)
    return top_recommendations.index.tolist()

# Örnek bir mentee için mentor önerileri al
recommendations = recommend_mentor(1, predictions_df)
print(f"Mentee 1 için önerilen mentorlar: {recommendations}")
```
