```python
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity
# Ilgi alanlarını ayrı öğeler olarak ele alarak her bir kullanıcı-öğe matrisini oluşturma
# Her bir kullanıcı (mentee) ve öğe (mentor) için birer vektör oluşturma
# Her bir vektörde, ilgi alanlarına göre bir sayısal değer bulma
interests_mentee = set()
interests_mentor = set()

for interest_list in merged_df['Mentee_Interests']:
    interests_mentee.update(interest_list.split(", "))

for interest_list in merged_df['Mentor_Interests']:
    interests_mentor.update(interest_list.split(", "))

interests_mentee = list(interests_mentee)
interests_mentor = list(interests_mentor)

interests_index_mentee = {interest: idx for idx, interest in enumerate(interests_mentee)}
interests_index_mentor = {interest: idx for idx, interest in enumerate(interests_mentor)}

def encode_interests(interest_list, interests_index):
    encoded = np.zeros(len(interests_index))
    for interest in interest_list.split(", "):
        encoded[interests_index[interest]] = 1
    return encoded

merged_df['encoded_mentee_interests'] = merged_df['Mentee_Interests'].apply(lambda x: encode_interests(x, interests_index_mentee))
merged_df['encoded_mentor_interests'] = merged_df['Mentor_Interests'].apply(lambda x: encode_interests(x, interests_index_mentor))

# NMF modelini başlatma ve uygulama
model = NMF(n_components=5, init='random', random_state=0)  # K = 5 olarak belirlendi
mentee_factors = model.fit_transform(list(merged_df['encoded_mentee_interests']))
mentor_factors = model.fit_transform(list(merged_df['encoded_mentor_interests']))

# Cosine benzerliğini hesaplama
similarity_matrix = cosine_similarity(mentee_factors, mentor_factors)
# Her mentee için en uygun N mentoru önerme
max_recommendations = 20  # Maksimum öneri sayısı

# Mentor önerilerini sınırlamak için bir fonksiyon
def limit_recommendations(recommendations, max_limit):
    return recommendations[:max_limit]

recommendations = []

for mentee_id, similarities in enumerate(similarity_matrix):
    # Benzerlik puanlarına göre sıralı indeksler
    mentor_indices = np.argsort(similarities)[::-1]
    
    # Mentee'nin kategorisini alma
    mentee_category = merged_df.iloc[mentee_id]['Mentee_Category']
    
    recommended_mentors = []  # Her bir mentee için önerilen mentorların listesi
    
    # Mentorları benzerlik puanına göre sırayla kontrol etme
    for mentor_index in mentor_indices:
        # Mentorun kategorisini ve ID'sini alma
        mentor_category = merged_df.iloc[mentor_index]['Mentor_Category']
        mentor_id = merged_df.iloc[mentor_index]['Mentor_ID']
        
        # Eğer mentorun kategorisi mentee'nin kategorisine eşitse ve daha önce öneri yapılmadıysa al
        if mentor_category == mentee_category and mentor_id not in recommended_mentors:
            recommended_mentors.append(mentor_id)
    
    # Öneri sayısını sınırlama
    recommended_mentors_limited = limit_recommendations(recommended_mentors, max_recommendations)
    
    # Sınırlı önerileri kaydetme
    for mentor_id in recommended_mentors_limited:
        recommendations.append((merged_df.iloc[mentee_id]['Mentee_ID'], mentor_id))

# Önerileri DataFrame olarak kaydetme
recommendation_df = pd.DataFrame(recommendations, columns=['Mentee_ID', 'Recommended_Mentor_ID'])
print(recommendation_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity

# Mentor ve mentee veri setlerini yükleme
mentor = pd.read_csv('mentor.csv')
mentee = pd.read_csv('mentee.csv')

# Mentor ve mentee veri setlerini birleştirme
merged_df = pd.merge(mentee, mentor, left_on='Mentor_ID', right_on='ID', suffixes=('_mentee', '_mentor'))

# Gerekli sütunları seçme
merged_df = merged_df[['ID_mentee', 'Category_mentee', 'Interests_mentee', 'Rating', 'ID_mentor', 'Category_mentor', 'Interests_mentor', 'Name_mentor', 'Surname_mentor']]

# Sütunları yeniden adlandırma
merged_df = merged_df.rename(columns={'ID_mentee': 'Mentee_ID', 'Category_mentee': 'Mentee_Category', 'Interests_mentee': 'Mentee_Interests',
                                      'ID_mentor': 'Mentor_ID', 'Category_mentor': 'Mentor_Category', 'Interests_mentor': 'Mentor_Interests',
                                      'Name_mentor': 'Mentor_Name', 'Surname_mentor': 'Mentor_Surname'})

# Ilgi alanlarını ayrı öğeler olarak ele alarak her bir kullanıcı-öğe matrisini oluşturma
# Her bir kullanıcı (mentee) ve öğe (mentor) için birer vektör oluşturulacak
# Her bir vektörde, ilgi alanlarına göre bir sayısal değer bulunacak
interests_mentee = set()
interests_mentor = set()

for interest_list in merged_df['Mentee_Interests']:
    interests_mentee.update(interest_list.split(", "))

for interest_list in merged_df['Mentor_Interests']:
    interests_mentor.update(interest_list.split(", "))

interests_mentee = list(interests_mentee)
interests_mentor = list(interests_mentor)

interests_index_mentee = {interest: idx for idx, interest in enumerate(interests_mentee)}
interests_index_mentor = {interest: idx for idx, interest in enumerate(interests_mentor)}

def encode_interests(interest_list, interests_index):
    encoded = np.zeros(len(interests_index))
    for interest in interest_list.split(", "):
        encoded[interests_index[interest]] = 1
    return encoded

merged_df['encoded_mentee_interests'] = merged_df['Mentee_Interests'].apply(lambda x: encode_interests(x, interests_index_mentee))
merged_df['encoded_mentor_interests'] = merged_df['Mentor_Interests'].apply(lambda x: encode_interests(x, interests_index_mentor))

# NMF modelini başlatma ve uygulama
model = NMF(n_components=5, init='random', random_state=0)  # K = 5 olarak belirlendi
mentee_factors = model.fit_transform(np.array(merged_df['encoded_mentee_interests'].tolist()))
mentor_factors = model.fit_transform(np.array(merged_df['encoded_mentor_interests'].tolist()))

# # NMF modelini başlatma ve uygulama
# model = NMF(n_components=len(interests_index_mentee), init='random', random_state=0)  # İlgi alanlarının sayısına göre ayarlandı
# mentee_factors = model.fit_transform(np.array(merged_df['encoded_mentee_interests'].tolist()))
# mentor_factors = model.transform(np.array(merged_df['encoded_mentor_interests'].tolist()))


# Cosine benzerliğini hesaplama
similarity_matrix = cosine_similarity(mentee_factors, mentor_factors)

# Her mentee için en uygun N mentoru önerme
N = 10  # Her mentee için önerilen mentor sayısı
recommendations = []

for mentee_id, similarities in enumerate(similarity_matrix):
    # Benzerlik puanlarına göre sıralı indeksler
    mentor_indices = np.argsort(similarities)[::-1]
    
    # Mentee'nin kategorisini al
    mentee_category = merged_df.iloc[mentee_id]['Mentee_Category']
    
    recommended_mentors = set()  # Her bir mentee için önerilen mentorların bir kümesi
    
    # Mentorları benzerlik puanına göre sırayla kontrol et
    for mentor_index in mentor_indices:
        # Eğer önerilen mentor sayısı N'e ulaştıysa döngüyü sonlandır
        if len(recommended_mentors) >= N:
            break
        
        # Mentorun kategorisini ve ID'sini al
        mentor_category = merged_df.iloc[mentor_index]['Mentor_Category']
        mentor_id = merged_df.iloc[mentor_index]['Mentor_ID']
        mentor_name = merged_df.iloc[mentor_index]['Mentor_Name']
        mentor_surname = merged_df.iloc[mentor_index]['Mentor_Surname']
        
        # Eğer mentorun kategorisi mentee'nin kategorisine eşitse ve daha önce öneri yapılmadıysa
        if mentor_category == mentee_category and mentor_id not in recommended_mentors:
            recommendations.append((merged_df.iloc[mentee_id]['Mentee_ID'], mentor_id, mentor_name, mentor_surname))
            recommended_mentors.add(mentor_id)

# Önerileri DataFrame olarak görüntüleme
recommendation_df2 = pd.DataFrame(recommendations, columns=['Mentee_ID', 'Recommended_Mentor_ID', 'Mentor_Name', 'Mentor_Surname'])
print(recommendation_df2)
```

```python
def make_custom_mentor_recommendations(new_mentee_interests, new_mentee_category, mentor_data, N=5):
    # Ilgi alanlarını ayrı öğeler olarak ele alarak yeni mentee'nin ilgi alanlarını kodlama
    encoded_new_mentee_interests = encode_interests(new_mentee_interests, interests_index_mentee)
    
    # NMF modeli ile yeni mentee'nin faktörlerini hesaplama
    new_mentee_factors = model.transform(np.array([encoded_new_mentee_interests]))
    
    # Cosine benzerliğini hesaplama
    similarities = cosine_similarity(new_mentee_factors, mentor_factors)[0]
    
    # Benzerlik puanlarına göre mentorları sıralama
    mentor_indices = np.argsort(similarities)[::-1]
    
    # Önerilen mentorları saklamak için bir liste oluşturma
    recommended_mentors = []
    
    # Her bir mentor için kategorisini ve ID'sini kontrol et
    for mentor_index in mentor_indices:
        mentor_category = mentor_data.iloc[mentor_index]['Category_mentor']
        mentor_id = mentor_data.iloc[mentor_index]['Mentor_ID']
        mentor_name = mentor_data.iloc[mentor_index]['Name_mentor']
        mentor_surname = mentor_data.iloc[mentor_index]['Surname_mentor']
        
        # Eğer mentorun kategorisi yeni mentee'nin kategorisine eşitse ve daha önce öneri yapılmadıysa
        if mentor_category == new_mentee_category and mentor_id not in recommended_mentors:
            recommended_mentors.append((mentor_id, mentor_name, mentor_surname))
            
        # Eğer önerilen mentor sayısı N'e ulaştıysa döngüyü sonlandır
        if len(recommended_mentors) >= N:
            break
    
    return recommended_mentors
```
