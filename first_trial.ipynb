```python
# Her bir ilgi alanı altında benzersiz subinterestleri bul
interest_subinterests = {}
for index, row in interests.iterrows():
    interest = row['Interest']
    if interest not in interest_subinterests:
        interest_subinterests[interest] = set()
    for column in interests.columns[1:]:
        subinterest = row[column]
        if subinterest != 'Belirtilmemiş':
            interest_subinterests[interest].add(subinterest)

# Her bir ilgi alanı altında benzersiz subinterestlerin sayısını hesapla
for interest, subinterest_set in interest_subinterests.items():
    print(f"Interest: {interest}")
    print("---------------------------------------------")
    for subinterest in subinterest_set:
        count = interests.iloc[:, 1:].apply(lambda x: x.eq(subinterest)).any(axis=1).sum()
        print(f"{subinterest}: {count}")
    print("\n")
```

```python
plt.figure(figsize=(12, 8))
for interest, subinterest_set in interest_subinterests.items():
    cleaned_subinterests = [str(subinterest) for subinterest in subinterest_set if pd.notnull(subinterest)]
    counts = [interests.iloc[:, 1:].apply(lambda x: x.eq(subinterest)).any(axis=1).sum() for subinterest in cleaned_subinterests]
    plt.bar(cleaned_subinterests, counts, alpha=0.7, label=interest)

plt.xlabel('Alt İlgi Alanı')
plt.ylabel('Mentor Sayısı')
plt.title('İlgi Alanlarına Göre Alt İlgi Alanları Dağılımı')
plt.xticks(rotation=45, ha='right')
plt.legend(title='İlgi Alanı', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Her bir ilgi alanı altında benzersiz subinterestleri bul
interest_subinterests = {}
for index, row in interests.iterrows():
    interest = row['Interest']
    if interest not in interest_subinterests:
        interest_subinterests[interest] = set()
    for column in interests.columns[1:]:
        subinterest = row[column]
        if pd.notnull(subinterest):  # NaN değerlerini kontrol et
            interest_subinterests[interest].add(subinterest)

# Her bir ilgi alanı altında benzersiz subinterestlerin sayısını hesapla
interest_counts = {}
for interest, subinterest_set in interest_subinterests.items():
    interest_counts[interest] = {}
    for subinterest in subinterest_set:
        count = interests.iloc[:, 1:].apply(lambda x: x.eq(subinterest)).any(axis=1).sum()
        interest_counts[interest][subinterest] = count

# Grafik için veri hazırlığı ve çubuk grafik oluşturma
fig, axs = plt.subplots(2, 5, figsize=(16, 8))  # 2 satır, 5 sütunlu bir alt grafik düzeni oluşturur
for i, (interest, subinterest_count) in enumerate(interest_counts.items()):
    ax = axs[i // 5, i % 5]  # i değerine göre uygun alt grafiğe erişim sağlar
    ax.bar(subinterest_count.keys(), subinterest_count.values(), color='purple')
    ax.set_title(f'{interest}', fontsize=12)
    ax.set_ylabel('Mentor Sayısı', fontsize=10)
    ax.tick_params(axis='x', rotation=90)  # x ekseninde etiketleri döndürme
plt.tight_layout()  # Grafikler arasındaki boşluğu otomatik olarak ayarlar
plt.show()
```

```python
#Eksik değerleri doldurma
for index, row in mentor.iterrows():
    interest = row['Interest']
    for i in range(2, 5):  # Subinterest sütunlarının indeksleri
        if pd.isnull(row[i]):  # NaN değer kontrolü
            # İlgili ilgi alanındaki diğer satırları al
            similar_rows = mentor[mentor['Interest'] == interest]
            # NaN olmayan değerlerin modunu al
            mode_value = similar_rows.iloc[:, i].mode().values[0]
            # NaN değerlerini mod ile doldur
            mentor.at[index, f'Subinterest_{i}'] = mode_value
```
